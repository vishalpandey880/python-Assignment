```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5])
y = [1,3,5,6,7]

unfoirm_dis = np.random.uniform(10, 20, 5)
normal_dis = np.random.normal(1,5,5)

plt.title("Data Distibution")
plt.xlabel("Data on x-axis")
plt.ylabel("Data on y-axis")
plt.grid(True)
plt.plot(unfoirm_dis, label ='Age')
plt.plot(normal_dis, label='Data')
plt.bar(x,y)
plt.savefig("my_graph.png")
plt.legend()



plt.show()
```
